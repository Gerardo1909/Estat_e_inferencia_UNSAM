Simular un conjunto de datos de tama√±o ùëõ = 10 a partir de una distribuci√≥n normal con ùúÉ = 5
y ùúé = 1. Graficar la likelihood. Usar el prior ùúÉ ‚àº N (0, 0.5) y calcular la distribuci√≥n posterior de
ùúÉ. Graficar las distribuciones prior y posterior.

### Desarrollo

Primero hago la generaci√≥n de los datos:

```python
from scipy.stats import norm

n_muestras = 10
datos = norm.rvs(loc=5, scale=1, size=n_muestras)
```

Ahora hago el gr√°fico de la likelihood:

```python
from matplotlib import pyplot as plt
import numpy as np

theta = np.linspace(-3, 10, 10)
likelihood = np.prod([norm.pdf(datos, loc=th, scale=1) for th in theta], axis=1)

plt.plot(theta, likelihood, label="Likelihood")
plt.xlabel("theta")
plt.ylabel("Likelihood")
plt.title("Likelihood de theta dados los datos")
plt.legend()
plt.show()
```

Ahora uso el prior propuesto para calcular la posterior de ùúÉ:

```python
prior = norm.pdf(theta, loc=0, scale=0.5)

posterior = likelihood * prior
posterior /= posterior.sum()  
```

Y finalmente grafico el prior y la posterior resultante:

```python
plt.plot(theta, prior, label="Prior")
plt.plot(theta, posterior, label="Posterior")
plt.xlabel("Theta")
plt.ylabel("Densidad")
plt.title("Prior y Posterior")
plt.legend()
plt.show()
```
