Particionar el intervalo [0, 1] en 100 intervalos de igual longitud. Considerar una matriz con
100 filas y 1000 columnas que en su fila enésima tiene los resultados de mil experimentos Bernoulli
con probabilidad que está en el intervalo enésimo.

## a)
Obtener un vector de largo 100 tomando la media de la matriz a lo largo del eje correspondiente y plotear ese vector contra el vector dado por los puntos medios de los intervalos en los que se particionó el [0, 1].

### Desarrollo

Primero armo la matriz que indica en el enunciado:

```python
import numpy as np

probabilidades = np.linspace(0, 1, 100)

matriz_resultados = np.zeros((100, 1000))

for i, prob in enumerate(probabilidades):
    matriz_resultados[i] = np.random.binomial(1, prob, 1000)
```

Y ahora armo el gráfico que me piden, mostrando **la relación entre el vector del probabilidades y la media de cada fila de la matriz**:

```python
import matplotlib.pyplot as plt

medias = matriz_resultados.mean(axis=1)
plt.figure(figsize=(8, 5))   
plt.plot(probabilidades, medias)
plt.title('Media de experimentos en cada fila en función de la probabilidad p asignada')
plt.xlabel('Probabilidad p de cada fila')
plt.ylabel('Media de experimentos en cada fila')
plt.show()
```
