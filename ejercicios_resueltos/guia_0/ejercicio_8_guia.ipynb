Para cada  entre 1 y 3000, generar observaciones 1, ..., ヰ de 1, ...,  v.a.i.i.d. con distribuci贸n exponencial con par谩metro  = 1/2, con  a elecci贸n. Luego, obtener la media de estas observaciones, es decir,

$$
\bar{x}_n = \frac{1}{n} \sum_{i=1}^{n} x_i.
$$

Graficar  vs ヂ e interpretar.


## Desarrollo

Primero importo la distribuci贸n exponencial de scipy y la inicializo con lambda igual a 1/2

```python
from scipy.stats import expon

exponencial = expon(scale=2)
```

Ahora defino una funci贸n que tome como argumento un n煤mero **** y que devuelva la media de la v.a con esas  observaciones:

```python
def gen_media_n(n: int, distribucion) -> float:
    # Genero n valores aleatorios de la distribuci贸n
    valores = distribucion.rvs(size=n)
    
    # Calculo la media muestral
    media = sum(valores) / n
    
    return media
```

Ahora genero las 3000 pruebas y las voy guardando en una lista para luego graficar

```python
medias_muestrales = [gen_media_n(n, exponencial) for n in range(1, 3001)]
```

Ahora finalmente genero el gr谩fico

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1,3001), medias_muestrales, label='Media Muestral')
plt.axhline(y=2, color='r', linestyle='--', label='Valor Esperado (2)')
plt.xlabel('n')
plt.ylabel('Media Muestral')
plt.title('n vs Media Muestral')
plt.legend()
plt.grid(True)
plt.show()
```
