Sea 洧녽 la cantidad de n칰meros 6 que salen en 1000 tiradas de un dado no necesariamente equilibrado, en donde 洧랚 es la probabilidad de sacar 6. Suponer un prior para 洧랚 definido de la siguiente manera:

**洧녞(洧랚 = 1/12) = 0.25 洧녞(洧랚 = 1/6) = 0.5 洧녞(洧랚 = 1/4) = 0.25**

## a)
Usando la aproximaci칩n normal de momentos obtener la distribuci칩n prior predictiva aproximada para y

### Desarrollo

En este caso me piden **洧녞(y)**, la cual representa la distribuci칩n **prior predictiva** para y. Esta se puede pensar por 
**ley de probabilidad total**, es decir que se multiplica cada probabilidad condicional de y dado uno de los posibles par치metros definidos en el 
prior multiplicado por la probabilidad puntual que tiene ese par치metro.

La forma de **洧녞(y)** ser칤a:

**洧녞(y) =  洧녞(y|洧랚 = 1/12)0.25 + 洧녞(y|洧랚 = 1/6)0.5 + 洧녞(y|洧랚 = 1/4)0.25**

Como estamos aproximando, cada **洧녞(y|洧랚 = valor)** seguir칤a una distribuci칩n normal, la cual tiene media y varianza 
seg칰n los momentos te칩ricos correspondientes para cada 洧랚, quedando de la forma:

**洧녞(y) =  Normal(250/3, 1375/18)0.25 + Normal(500/3, 1250/9)0.5 + Normal(250, 375/2)0.25**

Ahora paso a plantear todo esto en el c칩digo:

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return norm.pdf(x, loc=250/3, scale=np.sqrt(1375/18))*0.25 + norm.pdf(x, loc=500/3, scale=np.sqrt(1250/9))*0.5 + norm.pdf(x, loc=250, scale=np.sqrt(375/2))*0.25

# Genero un espacio de 10000 puntos entre -100 y 400
x = np.linspace(-500, 800, 10000)

# Calculo y
y = f(x)

# Hago el gr치fico de y
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Aproximaci칩n de binomial por m칠todo de momentos', fontsize=14)
plt.xlim(0,350)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
