# Ejercicio 6 - guía 8

Sea 𝑦 la cantidad de números 6 que salen en 1000 tiradas de un dado equilibrado.

## a)
Obtener la distribución de 𝑦 usando una aproximación normal de momentos (aproximar por
una normal a partir de la media y varianza poblacional).

### Desarrollo

En este caso tenemos que y sigue una distribución **Binomial(1000, 1/6)**, por lo tanto sabemos que sus momentos teóricos son de la forma:

**E(y) = 1000 * 1/6**

**Var(y) = 1000 * 1/6 * 5/6**

Entonces según nuestra aproximación tendríamos que y sigue **apróximadamente** una distribución **Normal(500/3 , 1250/9)**.

Ahora paso a plantear todo esto en el código:

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt

# Genero un espacio de 10000 puntos entre -100 y 400
x=np.linspace(0,1000,1000)

# Calculo la función de densidad de probabilidad de y Normal(500/3, 1250/9)
y = norm.pdf(x, 500/3, np.sqrt(1250/9))

# Hago un gráfico de la función de densidad de probabilidad
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', linestyle='-', linewidth=2)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.xlim(0,350)
plt.title('Aproximación de binomial por normal', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## b)
Dar los percentiles aproximados 5, 25, 50, 75 y 95 para la distribución de y.

```python
print(f'Percentil 5 de y: {norm.ppf(0.05, 500/3, 1250/9)}')
print(f'Percentil 25 de y: {norm.ppf(0.25, 500/3, 1250/9)}')
print(f'Percentil 50 de y: {norm.ppf(0.5, 500/3, 1250/9)}')
print(f'Percentil 75 de y: {norm.ppf(0.75, 500/3, 1250/9)}')
print(f'Percentil 95 de y: {norm.ppf(0.95, 500/3, 1250/9)}')
```
