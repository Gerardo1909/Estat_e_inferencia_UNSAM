# El rey markov que quería recorrer las 10 islas de su reino

La idea es utilizar **el algoritmo de metrópolis** para que el rey markov pueda pasar con más frecuencia por las islas que tienen mayor densidad de población, recordando
que hay 10 islas y que cada una tiene una población proporcional a su número correspondiente.

```python
import numpy as np 

num_semanas = int(1e5)
posiciones = np.zeros(num_semanas)
actual = 10

for i in range(num_semanas):
    # Registrar la posición actual
    posiciones[i] = actual

    # Lanzar moneda para generar propuesta
    propuesta = actual + np.random.choice([-1, 1])

    # Asegurarse de que haga bucles alrededor del archipiélago
    if propuesta < 1:
        propuesta = 10
    if propuesta > 10:
        propuesta = 1

    # ¿Moverse?
    prob_mover = propuesta / actual
    actual = propuesta if np.random.uniform() < prob_mover else actual
```

Ahora hago un histograma para verificar que el rey efectivamente estuvo más veces en las islas de mayor densidad:

```python
import matplotlib.pyplot as plt

# Generar histograma de las posiciones
plt.figure(figsize=(10, 6))
plt.hist(posiciones, bins=np.arange(0.5, 11.5, 1), edgecolor='black', align='mid')
plt.title('Histograma de las posiciones en las islas')
plt.xlabel('Isla')
plt.ylabel('Frecuencia')
plt.xticks(range(1, 11))
plt.show()
```
